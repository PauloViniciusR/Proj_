```python
import pandas as pd

arquivo = "base/base_cafe.xlsx"

mambo = pd.read_excel(arquivo, sheet_name="mambo")
marche = pd.read_excel(arquivo, sheet_name="marche")

mambo.head()
```

```python
mambo.columns
```

```python
colunas_importantes_mambo = [
    "SKU",
    "Titulo",
    "Fabricante",
    "Categoria",
    "Disponibilidade",
    "Preco",
    "Preço Oferta",
    "URL"
]

mambo = mambo[colunas_importantes_mambo]

mambo.head(50)
```

```python
mambo.info()
```

```python
mambo["Fabricante"].unique()
```

```python
mambo["Fabricante"].value_counts()
```

```python
mambo["Preco"] = pd.to_numeric(mambo["Preco"], errors="coerce")
mambo["Preço Oferta"] = pd.to_numeric(mambo["Preço Oferta"], errors="coerce") 
```

```python
mambo["Disponibilidade"].value_counts()
```

```python
mambo = mambo[mambo["Disponibilidade"] == "Disponivel"] 
mambo.head()
```
