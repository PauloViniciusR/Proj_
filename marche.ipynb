```python
import pandas as pd

arquivo = "base/base_cafe.xlsx"

mambo = pd.read_excel(arquivo, sheet_name="mambo")
marche = pd.read_excel(arquivo, sheet_name="marche")

marche.head()
```

```python
marche.columns
```

```python
colunas_importantes_marche = [
    "SKU",
    "Titulo",
    "Fabricante",
    "Categoria",
    "Disponibilidade",
    "Preco",
    "Preço Oferta",
    "URL"
]

marche = marche[colunas_importantes_marche]

marche.head()
```

```python
marche.info()
```

```python
marche["Fabricante"].unique()
```

```python
marche["Fabricante"].value_counts()
```

```python
marche["Preco"] = pd.to_numeric(marche["Preco"], errors="coerce")
marche["Preço Oferta"] = pd.to_numeric(marche["Preço Oferta"], errors="coerce") 
```

```python
marche["Disponibilidade"].value_counts()
```

```python
marche = marche[marche["Disponibilidade"] == "Disponivel"] 
marche.head()
```
